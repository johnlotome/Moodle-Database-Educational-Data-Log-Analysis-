#### Number of records in each of the tables given in the MIT section

```python
mit = ['mdl_logstore_standard_log','mdl_context','mdl_user','mdl_course', 'mdl_modules', 'mdl_course_modules',
        'mdl_course_modules_completion',  'mdl_grade_items','mdl_grade_grades', 'mdl_grade_categories',
        'mdl_grade_items_history','mdl_grade_grades_history','mdl_grade_categories_history', 'mdl_forum',
        'mdl_forum_discussions', 'mdl_forum_posts']
```

```python
def records_(tables_):
    total = %sql SELECT COUNT(*) as {tables_} FROM {tables_}
    return total
```

```python
for table in mit:
    print(records_(table))
```

#### Number of quiz submissions by hour of day

```python
%%sql
SELECT COUNT(id), EXTRACT(HOUR FROM to_timestamp(timecreated)) FROM mdl_logstore_standard_log 
WHERE action='submitted' AND component='mod_quiz' GROUP BY EXTRACT(HOUR FROM to_timestamp(timecreated)) LIMIT 10
```

#### Monthly usage time of learners who have confirmed and are not deleted

```python
%%sql 
SELECT COUNT(lastaccess - firstaccess) AS usagetime, EXTRACT (MONTH FROM to_timestamp(firstaccess)) AS month 
FROM mdl_user WHERE confirmed = 1 AND deleted = 0 GROUP BY EXTRACT (MONTH FROM to_timestamp(firstaccess))
```

#### Count of log events per user for the following verbs

```python
%%sql
SELECT * FROM mdl_logstore_standard_log LIMIT 3;
```

```python
verbs = ['loggedin', 'viewed', 'started', 'submitted', 'uploaded', 'updated',
         'searched', 'resumed', 'answered', 'attempted', 'abandoned']
```

```python
def actions_(verb_):
    results = %sql SELECT userid, COUNT(action) as action_counts FROM mdl_logstore_standard_log GROUP BY userid order by action_counts desc
    return results
```

```python
for actions in verbs:
    print(actions_(actions))
```
