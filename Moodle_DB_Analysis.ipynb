#### Monthly usage time of learners who have confirmed and are not deleted

```python
%%sql 
SELECT COUNT(lastaccess - firstaccess) AS usagetime, EXTRACT (MONTH FROM to_timestamp(firstaccess)) AS month 
FROM mdl_user WHERE confirmed = 1 AND deleted = 0 GROUP BY EXTRACT (MONTH FROM to_timestamp(firstaccess))
```

### Data Extraction Transformation and Loading (ETL)

```python
%%sql 
SELECT * FROM mdl_user LIMIT 5 
```

```python
%%sql
SELECT userid, action, timecreated
FROM mdl_logstore_standard_log 
WHERE userid = 917 AND action = 'started'
GROUP BY userid, action, timecreated
LIMIT 20
```

```python
class Dedication:
    def __init__(self, database = 'moodle', user = 'lotomej'):
        self.database = database
        self.user = user
        self.conn = psycopg2.connect(host="localhost",password=1234, port = 5432, database=self.database, user=self.user)
    def get_table(self):
        self.activity = pd.read_sql_query("select * from mdl_logstore_standard_log", con = self.conn)
        return self.activity
    def get_dedications(self, user):
        table = activity
        df = table.loc[activity.userid == user]
        logins = df.loc[df.action == 'loggedin']
        others = df.loc[df.action != 'loggedin']
        login_stamps = logins.timecreated.to_list()
        other_stamps = others.timecreated.to_list() 
        latest_time = max(other_stamps) 
        last_events = []
        for i in range(1,len(login_stamps)):
            try:
                latest_before_login = max(dt for dt in other_stamps if dt < login_stamps[i])
                 
            except Exception:
                latest_before_login = login_stamps[i-1] 
            last_events.append(latest_before_login)
        latest_event = [latest_time]
        last_events = last_events + latest_event
        times = []
        for i,j in zip(last_events, login_stamps):
            diff = i - j
            diff = diff if diff < 7200 else 7200  
            times.append(diff)
        try:
            user_dedication_total = sum(times)
            user_dedication_total = user_dedication_total if user_dedication_total >= 0 else 0 
        except Exception: 
            user_dedication_total = 0
        return user_dedication_total
    def reconstruct_table(self, totals_list):
        user_ids = activity.userid.unique().tolist()
        dedication_time_totals = totals_list
        df = pd.DataFrame(columns = ['userid', 'dedication_time'])
        df['userid'] = user_ids
        df['dedication_time'] = dedication_time_totals
        return df
    conn.close()
             
```

```python
d = Dedication()
activity = d.get_table()
dedication_totals = []
for user in activity.userid.unique().tolist():
    total = d.get_dedications(user)
    dedication_totals.append(total)
len(dedication_totals) 
```

```python
dedication_ = d.reconstruct_table(dedication_totals)
dedication_.head()
```

```python
def top_percentage(df, percent):
    '''
        takes the dataframe, the percent to groupby
    '''
    total_len = df.shape[0]
    top = int((total_len * percent)/100)
    return df.iloc[:top,]
```

```python
top_percentage(dedication_, 25)
```
